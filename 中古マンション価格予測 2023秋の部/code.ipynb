```python
x = df_train.drop('取引価格（総額）_log',axis=1)
x
```

```python
y = df_train['取引価格（総額）_log']
y
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
x_train, x_test, t_train, t_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
lgb_train = lgb.Dataset(x_train, t_train)
lgb_eval = lgb.Dataset(x_test, t_test)
```

```python
#light gbmモデル構築
params = {
    'objective':'regression',
    'seed':0,
    'learning_rate':0.1,
    'num_iteration':3250,
    # 'verbose':0,
    # 'max_depth':20,
    'metrics':'neg_mean_absolute_error',
    'num_leaves':60,
    # 'bagging_freq':1,
    # 'bagging_fraction':0.2,
    # 'min_data_in_leaf':20,
}

model = lgb.train(
    params,
    lgb_train,
    valid_sets = lgb_eval,
    num_boost_round = 50,
    # early_stopping_round=10,
)
```

```python
pre = model.predict(x_test)
pre
```

```python
plt.scatter(t_test, pre, s=20, c=None, marker='o', cmap=None, norm=None,
                   vmin=None, vmax=None, alpha=None, linewidths=None,
                   edgecolors=None, data=None)

plt.plot([min(t_test), max(t_test)], [min(pre), max(pre)], linestyle='--', color='red')
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(t_test, pre)
mae
```
